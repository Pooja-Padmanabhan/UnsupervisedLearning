### B. Plot a visual and find elbow point

```python
# Plot the graph to visualize the optimal number of cluster through the Elbow Method
X = merged_dropped.values

# Define the range of clusters to consider
min_clusters = 2
max_clusters = 10

# Initialize lists to store the inertia values (within-cluster sum of squares) for different clusters
inertia_values = []

# Perform K-Means clustering for different numbers of clusters
for n_clusters in range(min_clusters, max_clusters+1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Plot the inertia values to determine the optimal number of clusters
plt.plot(range(min_clusters, max_clusters+1), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Car Attributes')
plt.title('Car Attributes vs. Number of Clusters')
plt.show()
```

### C. On the above visual, highlight which are the possible Elbow points.

```python
X = merged_dropped.values

# Define the range of clusters to consider
min_clusters = 2
max_clusters = 10

# Initialize lists to store the inertia values (within-cluster sum of squares) for different clusters
inertia_values = []

# Perform K-Means clustering for different numbers of clusters
for n_clusters in range(min_clusters, max_clusters+1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Plot the inertia values to determine the optimal number of clusters
plt.plot(range(min_clusters, max_clusters+1), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Inertia vs. Number of Clusters')

# Find the possible elbow points
possible_elbows = []
for i in range(1, len(inertia_values) - 1):
    prev_slope = inertia_values[i - 1] - inertia_values[i]
    next_slope = inertia_values[i] - inertia_values[i + 1]
    if prev_slope >= next_slope:
        possible_elbows.append(i + min_clusters)

# Highlight the possible elbow points on the plot
for elbow in possible_elbows:
    plt.scatter(elbow, inertia_values[elbow - min_clusters], color='red', marker='o')
    plt.annotate(f'Elbow {elbow}', (elbow, inertia_values[elbow - min_clusters]),
                 textcoords="offset points", xytext=(-10,10), ha='center', color='red')

plt.show()

# Print the possible elbow points
print("Possible elbow point(s):", possible_elbows)
```

### 5. Data Understanding & Cleaning:
### A. Explain pre-requisite/assumptions of PCA.

Principal Component Analysis (PCA) is a dimensionality reduction technique that is commonly used to analyze and visualize high-dimensional data. Before applying PCA, there are a few pre-requisites and assumptions to consider:

1. Numeric Data: PCA is designed to work with numeric data, such as continuous variables. Categorical or textual data typically need to be preprocessed and transformed into numerical representations before applying PCA.

2. Linear Relationships: PCA assumes that the relationships between variables are linear. It seeks to find linear combinations of the original variables that capture the most significant variability in the data.

3. Standardization: It is generally recommended to standardize the variables before applying PCA. Standardization involves scaling the variables to have zero mean and unit variance. This step ensures that variables with larger variances do not dominate the PCA results. Standardization is particularly important when the variables have different scales or units.

4. Large Variability: PCA tends to work best when there is substantial variability in the dataset. Variables that exhibit little variability or have near-zero variance may not contribute much to the principal components and may be less informative.

5. Independence: PCA assumes that the variables are independent or have weak correlations. If the variables are highly correlated, PCA may not effectively capture the underlying structure of the data. In such cases, other techniques like Factor Analysis or Independent Component Analysis (ICA) may be more appropriate.

6. Linearity and Outliers: PCA assumes that the data points and variables follow linear patterns. Non-linear relationships or the presence of outliers may affect the performance of PCA. Outliers can disproportionately influence the principal components and distort the results. Therefore, it's important to preprocess the data and handle outliers appropriately before applying PCA.

7. Large Sample Size: PCA performs better with larger sample sizes. As the sample size increases, the estimates of the principal components become more reliable and stable. If the sample size is small, the results of PCA may be less robust and subject to overfitting.

It's important to consider these pre-requisites and assumptions while applying PCA to ensure the reliability and interpretability of the results. Additionally, it's recommended to explore and understand the characteristics of the dataset before applying PCA to assess its suitability and address any specific requirements or constraints.

### B. Explain advantages and limitations of PCA.

#### Advantages of PCA:

1. Dimensionality Reduction: PCA is primarily used for dimensionality reduction. It reduces the number of variables while preserving the most important information. By selecting a subset of the principal components that explain most of the variance, PCA allows for simplifying complex datasets and visualizing high-dimensional data in a lower-dimensional space.

2. Feature Extraction: PCA enables the extraction of a set of orthogonal variables (principal components) that are linear combinations of the original variables. These principal components are sorted based on the amount of variance they explain. They can be used as new features that capture the most significant patterns and variations in the data.

3. Data Visualization: PCA facilitates data visualization by reducing the data to two or three dimensions. It helps in understanding the structure and relationships between variables, identifying clusters or patterns, and visually exploring the data in a more interpretable format. This is particularly useful for exploratory data analysis and communicating the insights to others.

4. Multicollinearity Handling: PCA can address the issue of multicollinearity, which occurs when the variables in a dataset are highly correlated. It transforms the original correlated variables into a set of uncorrelated principal components, eliminating the problem of multicollinearity. This is beneficial in regression analysis and other modeling techniques that assume independent variables.

#### Limitations of PCA:

1. Linearity Assumption: PCA assumes that the relationships between variables are linear. If the data has non-linear relationships, PCA may not effectively capture the underlying structure. In such cases, nonlinear dimensionality reduction techniques like Kernel PCA or t-SNE may be more appropriate.

2. Information Loss: Although PCA aims to preserve the most important information, there is generally some loss of information during dimensionality reduction. Lower-dimensional representations may not fully capture all the details present in the original data. It's important to consider the trade-off between dimensionality reduction and preserving the desired level of information.

3. Interpretability of Components: While PCA provides a reduced set of variables (principal components), interpreting these components in the context of the original variables may not always be straightforward. The interpretation of the components depends on the domain knowledge and the specific context of the data.

4. Sensitivity to Outliers: PCA is sensitive to outliers since outliers can disproportionately influence the calculation of principal components. Outliers may distort the results and lead to suboptimal dimensionality reduction. It's important to preprocess the data and handle outliers appropriately before applying PCA.

5. Subjectivity in Component Selection: Selecting the number of principal components to retain is subjective and may require additional considerations. The choice of the number of components impacts the amount of variance explained and the trade-off between dimensionality reduction and information loss. Different criteria such as cumulative variance explained or scree plots can be used, but the selection process may still involve some level of subjectivity.

6. Computational Complexity: PCA can be computationally expensive for large datasets, especially when the number of variables is large. The computational complexity increases with the size of the dataset and the number of variables. Efficient algorithms and implementations are available, but the scalability should be considered for very large datasets.
